```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms import bipartite
```

```python
G = nx.davis_southern_women_graph()
women = G.graph["top"]
clubs = G.graph["bottom"]

print(women)
print(clubs)
```

```python
for node in G.nodes(data=True):
    print(node)
```

```python
print("Biadjacency matrix")
print(bipartite.biadjacency_matrix(G, women, clubs))
```

```python
pos = nx.spring_layout(G, seed=648)  # Seed layout for reproducible node positions
nx.draw(G, pos)
plt.show()
```

#### **Projecting women onto network G**
When we project women onto the total network we see that the lady with the most friends with multiple meetings is Theresa Anderson

```python
# project bipartite graph onto women nodes keeping number of co-occurrence
# the degree computed is weighted and counts the total number of shared contacts
W = bipartite.weighted_projected_graph(G, women)
print()
print("#Friend meetings, Member")
for w in women:
    print(f"{W.degree(w, weight='weight')} {w}")

```

```python
for node in W.nodes(data=True):
    print(node)
```

```python
edge_labels1 = nx.get_edge_attributes(W, 'weight')
pos1 = nx.spring_layout(W, seed=648)  # Seed layout for reproducible node positions
nx.draw(W, pos1, with_labels=True)
nx.draw_networkx_edge_labels(W, pos1, edge_labels=edge_labels1)
plt.show()
```

#### **Projecting club onto network G**
When we project clubs onto the total network we see that the most popular club per women attendance is E8.

```python
# project bipartite graph onto club nodes keeping number of co-occurrence
# the degree computed is weighted and counts the total number of shared contacts
X = bipartite.weighted_projected_graph(G, clubs)
print()
print("#Members, Clubs")
for x in clubs:
    print(f"{X.degree(x, weight='weight')} {x}")

```

```python
for node in X.nodes(data=True):
    print(node)
```

```python
edge_labels2 = nx.get_edge_attributes(X, 'weight')
pos2 = nx.spring_layout(X, seed=648)  # Seed layout for reproducible node positions
nx.draw(X, pos2, with_labels=True)
nx.draw_networkx_edge_labels(X, pos2, edge_labels=edge_labels2)
plt.show()
```
